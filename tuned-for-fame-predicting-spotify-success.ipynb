```python
# Drop columns that are not useful for ML
df.drop(['Unnamed: 0', 'track_id', 'album_name', 'track_name'], axis=1, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
# Create a new binary target column: 1 = popular, 0 = not popular
df['popular'] = df['popularity'].apply(lambda x: 1 if x >= 60 else 0)

# Check how many popular vs not-popular songs
df['popular'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: improve chart styles
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='track_genre', hue='popular', order=df['track_genre'].value_counts().index[:10])
plt.xticks(rotation=45)
plt.title("Popularity by Track Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.legend(title="Popular")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='danceability', y='energy', hue='popular', alpha=0.5)
plt.title("Energy vs Danceability (Popular vs Not)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='popular', y='valence')
plt.title("Valence Distribution by Popularity")
plt.xlabel("Is Popular")
plt.ylabel("Valence (Happiness)")
plt.show()
```
